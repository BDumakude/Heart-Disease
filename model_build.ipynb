```python
from sklearn.metrics import classification_report
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.svm import SVC 
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression 
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
```

```python
class CustomMinMax(BaseEstimator, TransformerMixin):
    def __init__(self, columns: list) -> None:
        self.scaler = MinMaxScaler()
        self.columns = columns
    
    def fit(self, X, y=None):
        self.scaler.fit(X[self.columns])
        return self
    
    def transform(self, X):
        initial_column_order = X.columns
        X_scaled = pd.DataFrame(self.scaler.transform(X[self.columns]), columns=self.columns, index=X.index)
        X_not_scaled = X.drop(self.columns, axis=1)
        return pd.concat([X_scaled, X_not_scaled], axis=1)[initial_column_order]
```

```python
X = df_input.drop('target', axis=1)
y = df_input['target']
```

```python
columns_to_scale = []
for col in df_input.columns[:-1]:
    if df_input[col].value_counts().shape[0] > 5:
        columns_to_scale.append(col)
print(columns_to_scale)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify=y)
```

```python
pipe1 = Pipeline([('scaler', CustomMinMax(columns=columns_to_scale)), ('svc', SVC())])
pipe2 = Pipeline([('scaler', CustomMinMax(columns=columns_to_scale)), ('logreg', LogisticRegression())])

param_grid1 = {'svc__C': [0.001, 0.01, 0.1, 1, 10, 100],
               'svc__gamma': [0.001, 0.01, 0.1, 1, 10, 100]}
prama_grid2 = {'logreg__C': [0.001, 0.01, 0.1, 1, 10, 100]}

grid1 = GridSearchCV(pipe1, param_grid=param_grid1, cv=5)
grid2 = GridSearchCV(pipe2, param_grid=prama_grid2, cv=5)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
grid1.fit(X_train, y_train)
grid2.fit(X_train, y_train)

print('=============SVC==============')
print('Best cross-validation accuracy', grid1.best_score_)
print('Best test accuracy', grid1.score(X_test, y_test))
print('Best parameters', grid1.best_params_)
print('==============================')

print('=============LogReg==============')
print('Best cross-validation accuracy', grid2.best_score_)
print('Best test accuracy', grid2.score(X_test, y_test))
print('Best parameters', grid2.best_params_)
print('==============================')
```

```python
best_model = grid1.best_estimator_
y_pred = best_model.predict(X_test)
report = classification_report(y_test, y_pred)
print(report)
```
